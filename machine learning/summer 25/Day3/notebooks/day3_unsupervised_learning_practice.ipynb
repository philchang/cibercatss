Welcome to Day 3! Today we'll explore unsupervised learning algorithms with hands-on coding. We'll focus on **dimensionality reduction techniques**: PCA and t-SNE.

```python
##  1. Setup and Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

%matplotlib inline
sns.set(style='whitegrid')
```

```python
##  2. Load and Explore Dataset
#We'll use the classic **Iris dataset**, which contains 4 features and 3 species of flowers.


iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name="species")

print(X.head())
print("Target Classes:", iris.target_names)
```

```python
## 3. Standardize the Data
#Most unsupervised learning techniques (like PCA and t-SNE) require features to be on the same scale.


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
## 4. Principal Component Analysis (PCA)
#Reduce from 4D â†’ 2D and visualize the structure.


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pca_df["species"] = y

plt.figure(figsize=(8, 6))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="species", palette="Set2")
plt.title("PCA - 2D Projection of Iris Dataset")
plt.show()

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

## 5. t-Distributed Stochastic Neighbor Embedding (t-SNE)

```python


tsne = TSNE(n_components=2, perplexity=30, learning_rate=200, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

tsne_df = pd.DataFrame(X_tsne, columns=["Dim1", "Dim2"])
tsne_df["species"] = y

plt.figure(figsize=(8, 6))
sns.scatterplot(data=tsne_df, x="Dim1", y="Dim2", hue="species", palette="Set1")
plt.title("t-SNE - 2D Projection of Iris Dataset")
plt.show()
```
