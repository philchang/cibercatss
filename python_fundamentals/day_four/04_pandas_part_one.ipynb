* The reason `df['Index']` doesn't work is because the index values are not treated as a separate column in the DataFrame. 

* Instead, they are used as the row labels to identify and access specific rows in the DataFrame.

* By default, the index values are not considered a column in the DataFrame and are not directly accessible using standard column indexing syntax. However, you can access the index indirectly using the `.index` attribute.

If you want to access the index as a column in the DataFrame, you can reset the index using the `reset_index()` method, which will add the current index as a column and assign a new default integer index to the DataFrame

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'Emily'],
        'Age': [25, 30, 35, 28],
        'Country': ['USA', 'Canada', 'UK', 'Australia']}
df = pd.DataFrame(data)

# Resetting the index
df_with_index_column = df.reset_index()

# Displaying the DataFrame with the index as a column
print("DataFrame with Index as a Column:")
print(df_with_index_column)
print()

# Accessing the 'Index' column
index_column = df_with_index_column['index']

print(index_column)
```

# Difference between loc and iloc

## Key differences:

1. Indexing Method:

* loc: Label-based indexing.
* iloc: Integer-based indexing.

2. Indexer Type:

* loc: Accepts label-based indexers (labels or boolean arrays).
* iloc: Accepts integer-based indexers (integers, lists/arrays of integers, or slices).

3. Inclusive Range:

* loc: Inclusive of the endpoints.
* iloc: Exclusive of the endpoint of a slice (similar to Python slicing).

4. Usage of Index Labels:

* loc: Uses index labels to select rows and columns.
* iloc: Ignores index labels and uses the integer position to select rows and columns.

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'Emily'],
        'Age': [25, 30, 35, 28],
        'Country': ['USA', 'Canada', 'UK', 'Australia']}
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D'])

# Using loc to select based on labels
print("Using loc:")
print(df.loc['B', 'Age'])  # Selects the value at row 'B' and column 'Age'
print(df.loc['A':'C', 'Name':'Country'])  # Selects rows 'A' to 'C' and columns 'Name' to 'Country'
print()

# Using iloc to select based on integer positions
print("Using iloc:")
print(df.iloc[1, 1])  # Selects the value at row 1 and column 1
print(df.iloc[0:3, 0:2])  # Selects rows 0 to 2 (exclusive) and columns 0 to 1
```

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'Emily'],
        'Age': [25, 30, 35, 28],
        'Country': ['USA', 'Canada', 'UK', 'Australia']}
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D'])

# Using loc with label-based indexers
print("Using loc:")
print(df.loc[['A', 'C'], 'Age'])  # Selects the 'Age' column for rows 'A' and 'C'
print(df.loc[df['Age'] > 30])  # Selects rows where Age is greater than 30
print()

# Using iloc with integer-based indexers
print("Using iloc:")
print(df.iloc[0:2, 1])  # Selects the second column for rows 0 to 1
print(df.iloc[[True, False, True, False]])  # Selects rows with boolean indexer
print()
```

`iloc` and `loc` are both helpful in different situations, depending on the type of indexing or selection you need to perform on your data. Here are some scenarios where iloc and loc can be particularly useful:

1. **iloc**:

* When you want to access data based on integer-based positional indexing.
* When you need to perform slicing or indexing operations using integer positions rather than labels.
* When you want to extract specific rows or columns based on their integer positions.
* When you want to perform calculations or operations on a subset of the DataFrame using integer-based indexing.

2. **loc**: 

* When you want to access data based on label-based indexing.
* When you need to perform slicing or indexing operations using label-based indexing.
* When you want to extract specific rows or columns based on their labels.
* When you want to perform calculations or operations on a subset of the DataFrame using label-based indexing.



```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'Emily'],
        'Age': [25, 30, 35, 28],
        'Country': ['USA', 'Canada', 'UK', 'Australia']}
df = pd.DataFrame(data)

# Selecting specific rows using iloc
subset = df.iloc[1:3, :]  # Selects rows 1 and 2, all columns
print(subset)
```

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'Emily'],
        'Age': [25, 30, 35, 28],
        'Country': ['USA', 'Canada', 'UK', 'Australia']}
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D'])

# Selecting specific columns using loc
subset = df.loc[:, ['Name', 'Country']]  # Selects all rows, 'Name' and 'Country' columns
print(subset)
```

# Selection and Filtering 

## Example 1: Selecting specific columns

```python
# Selecting the 'Name' and 'Age' columns
selected_columns = df[['Name', 'Age']]
print(selected_columns)
```

## Example 2: Filtering rows based on a condition

```python
# Filtering rows where Age is greater than 30
filtered_rows = df[df['Age'] > 30]
print(filtered_rows)
```

## Example 3: Selecting rows based on multiple conditions

```python
# Selecting rows where Age is between 25 and 35 and Country is either 'USA' or 'Canada'
selected_rows = df[(df['Age'] >= 25) & (df['Age'] <= 35) 
                  & ((df['Country'] == 'USA') 
                  | (df['Country'] == 'Canada'))]
print(selected_rows)
```

# Data Aggregation

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['John', 'Alice', 'Bob', 'Emily', 'Johnson'],
        'Age': [25, 30, 35, 28, 22],
        'Country': ['USA', 'Canada', 'UK', 'Australia', 'USA'],
        'Salary': [50000, 60000, 70000, 55000, 45000]}
df = pd.DataFrame(data)




df.head()
```

```python
# Aggregating data using groupby
grouped_data = df.groupby('Country')
grouped_data.head()
```

```python
# Calculating the mean of salaries by country
mean_salary_by_country = grouped_data['Salary'].mean()
# Displaying the aggregated results
print("Mean Salary by Country:")
print(mean_salary_by_country)
print()


```

```python
# Calculating the maximum age by country
max_age_by_country = grouped_data['Age'].max()

print("Maximum Age by Country:")
print(max_age_by_country)
print()


```

```python
# Calculating the count of names by country
name_count_by_country = grouped_data['Name'].count()

print("Count of Names by Country:")
print(name_count_by_country)
```

# Data Cleaning and Transformation

```python
import pandas as pd

# Creating a DataFrame with dirty data
data = {'First_Name': ['John', 'Alice', 'Bob', 'Emily', 'John'],
        'Last_Name': ['Doe', 'Smith', 'Johnson', 'Doe', 'Smith'],
        'Age': ['25', '30', '35', '28', '22'],
        'Salary': ['50000', '60000', '70000', '55000', '45000']}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Renaming columns
df.rename(columns={'First_Name': 'First Name', 'Last_Name': 'Last Name'}, inplace=True)
```

```python
df.head()
```

```python
# Sorting data by Age in descending order
df.sort_values(by='Age', ascending=False, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# Converting Age and Salary to numeric data types
df['Age'] = pd.to_numeric(df['Age'])
df['Salary'] = pd.to_numeric(df['Salary'])
```

```python
df.dtypes
```

```python
# Applying a function to elements of Age column
df['Age'] = df['Age'].apply(lambda x: x + 1)
```

```python
df.head()
```

```python
# Merging/joining multiple DataFrames
additional_data = {'First Name': ['Sam', 'Ella'],
                   'Last Name': ['Jones', 'Williams'],
                   'Age': [32, 26],
                   'Salary': [65000, 55000]}
df_additional = pd.DataFrame(additional_data)
df_additional.head()

```

```python
pd.concat([df, df_additional], axis=0, ignore_index=True)
```

```python
# Concatenate the DataFrames row-wise
concatenated_df = pd.concat([df, df_additional], axis=0, ignore_index=True)

print(concatenated_df)
```

# Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a DataFrame
data = {'Country': ['USA', 'Canada', 'UK', 'Australia'],
        'Population': [328, 38, 67, 25],
        'GDP': [21427, 1780, 2826, 1363]}
df = pd.DataFrame(data)

# Plotting a bar chart using the plot method in Pandas
df.plot(x='Country', y='Population', kind='bar', title='Population by Country')

# Plotting a scatter plot using Matplotlib
plt.figure()
plt.scatter(df['GDP'], df['Population'])
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Population vs. GDP')



# Displaying the plots
plt.show()
```
